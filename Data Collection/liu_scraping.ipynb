```python
import pdfplumber

# List of PDF file paths
pdf_files = [
    "/content/CHED-POS (1).pdf",
    "/content/TEBC-POS.pdf",
    "/content/TESL-POS.pdf",
    "/content/TEPM-POS.pdf",
    "/content/TRNS-POS.pdf"
]

for pdf_file in pdf_files:
    titles = []  

    with pdfplumber.open(pdf_file) as pdf:
        for page in pdf.pages:
            tables = page.extract_tables()
            for table in tables:
                for row in table:
                    if row and len(row) > 1:
                        title = row[1]
                        if title and title.lower() != "title" and len(title.strip()) > 3:
                            titles.append(title.strip())

    
    titles = list(dict.fromkeys(titles))

    print(f"\nðŸ“„ Titles from: {pdf_file}\n")
    for title in titles:
        print(title)

    print(f"\nâœ… Done: {pdf_file}\n{'-'*50}")
```
